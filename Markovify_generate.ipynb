# Markov chain generates text
 
 Code reference: Week5
 
 The ‘state_size’ parameter of the Markov chain in this model is set to 2, so that the generative model will consider the context of the current word and the previous word to generate the next word.
 
 ### Import libraries
 
 Import the Markovify library for text generation：

```python
import os
import markovify
```

###  Function definitions

Read the contents of the text file and store it in a corpus variable.

```python
# read the file's content into the corpus variable
def load_single_text_file(path):
    with open(path, 'r', encoding='utf-8') as f:
        corpus = f.read()
        return corpus
    
data_path = 'data/Troye Sivan_Lyrics.txt'
corpus = load_single_text_file(data_path)

# Split the corpus into words and represent each word as a list of characters
corpus_char_list = [list(word) for word in corpus]
```

### Generate lyrics

Creating a text model using Markovify and generating text.


```python
# Create a Markovify text model
text_model = markovify.Text(corpus, state_size=2) 

# Create an empty string new_str to store the generated text
# Generate 150 short sentences with a maximum of 30 characters and add them to new_str
new_str = ''
for i in range(150):
    example_output = text_model.make_short_sentence(max_chars=30)  
    if example_output:
        new_str += example_output + ' \n' 

print(new_str)
```

### Save file

Save the generated text to a file.

```python
try:
    f = open("New lyrics/markovify_lyrics.txt","a")
except:
    f = open("New lyrics/markovify_lyrics.txt","w")
f.write(new_str)
f.close()
```
