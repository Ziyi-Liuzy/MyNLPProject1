The purpose of this code is to load lyrics data from a JSON file and then write these lyrics to a text file, for the purpose of saving and further processing.

```python
# loading the data and deleting the original file
with open('Lyrics_TroyeSivan.json') as json_data:
    data = json.load(json_data)
os.remove('Lyrics_TroyeSivan.json')

# Create an empty list 'lyrics' to store the lyrics data
lyrics=[]
for song in data['songs']:
    # Extract the lyrics of each song and append them to the 'lyrics' list
    lyrics.append(song['lyrics'])

# Save lyrics to a text file
filepath = artist_name+"_Lyrics.txt"

f = open(filepath,'w')

for song in lyrics:
    f.write(str(song))
f.close()
```

## Processing Lyrics Dataset

Pre-process the dataset to remove content that is not part of the lyrics and initially standardise the format of the text file.

```python
with open('Troye Sivan_Lyrics.txt', 'r') as file:
    lines = file.readlines()

cleaned_lines = []

for line in lines:
    # Remove leading and trailing whitespace and check if it's an empty line
    stripped_line = line.strip()
    if stripped_line:
        # Check if it does not contain '[]' and '()', and if not, add it to the new list
        if '[' not in stripped_line and ']' not in stripped_line and '(' not in stripped_line and ')' not in stripped_line:
            cleaned_lines.append(stripped_line)

# Write the processed text back to a file
with open('Troye Sivan_Lyrics.txt', 'w') as file:
    file.write('\n'.join(cleaned_lines))
```
